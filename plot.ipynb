```python
import pandas as pd
game = pd.read_csv("game.csv")

# show how many turns taken for a player to win a bot
bot = game[(game["Player 2"] == "Bot")] 

bot_game=bot[["Player 1", "Turns taken"]]
bot_game.plot.scatter(x = "Player 1", y = "Turns taken")
```

```python
# show a histogram of how many turns is taken to win in a human game. 
human_game=game[(game["Player 2"] != "Bot")] 
human_game.plot(kind = 'hist', y = "Turns taken")
```

```python
# shows who has played with whom
game.plot.scatter(x = "Player 1", y = "Player 2")
```
