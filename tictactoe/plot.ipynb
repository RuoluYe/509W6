```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
game = pd.read_csv("game.csv")

# show how many turns taken for a player to win a bot
bot = game[(game["Player 2"] == "Bot")] 

bot_game=bot[["Player 1", "Turns taken"]]
fig = bot_game.plot.scatter(x = "Player 1", y = "Turns taken")
fig.figure.savefig('.\images\winbot.png')
```

```python
# show a histogram of how many turns is taken to win in a human game. 
human_game=game[(game["Player 2"] != "Bot")] 
fig = human_game.plot(kind = 'hist', y = "Turns taken")
fig.figure.savefig('.\images\human game.png')
```

```python
# shows who has played with whom
fig = game.plot.scatter(x = "Player 1", y = "Player 2")
fig.figure.savefig('.\images\playerStat.png')
```
